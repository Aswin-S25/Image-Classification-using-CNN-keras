# Image Classification using CNN

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
```

## Load Dataset

```python
X_train = np.loadtxt('dataset/input.csv', delimiter=',')
Y_train = np.loadtxt('dataset/labels.csv', delimiter=',')

X_test = np.loadtxt('dataset/input_test.csv', delimiter=',')
Y_test = np.loadtxt('dataset/labels_test.csv', delimiter=',')
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of Y_train: ", Y_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of Y_test: ", Y_test.shape)
```

## Reshape the dataset

```python
X_train = X_train.reshape(len(X_train), 100, 100, 3)
Y_train = Y_train.reshape(len(Y_train), 1)

X_test = X_test.reshape(len(X_test), 100, 100, 3)
Y_test = Y_test.reshape(len(Y_test), 1)
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of Y_train: ", Y_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of Y_test: ", Y_test.shape)
```

## Normalization of dataset

```python
X_train[1, :]
```

```python
X_train = X_train/255
X_test = X_test/255
```

## loading sample images

```python
idx = random.randint(0, len(X_train))
plt.imshow(X_train[idx, :])
plt.show()
```

## Model

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),  # Adjusted input shape
    MaxPooling2D((2, 2), padding='same'),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss= 'binary_crossentropy', optimizer='adam', metrics= ['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs = 10, batch_size=64)
```

```python
model.evaluate(X_test, Y_test)
```

## Making Prediction

```python
idx2 = random.randint(0, len(Y_test))
plt.imshow(X_test[idx2, :])
plt.show()

y_pred = model.predict(X_test[idx2, :].reshape(1, 100, 100, 3))
y_pred = y_pred > 0.5

if(y_pred == 0):
    pred = 'dog'
else:
    pred = 'cat'

print("our model says it is", pred)
```
